```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
```

```python
X = data.drop(columns=['Unnamed: 0', 'class'])
y = data['class']
```

```python
# Encode categorical columns
categorical_cols = X.select_dtypes(include='object').columns
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)
```

```python
# Encode target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded
)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Random Forest Classification
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
```

```python
# Metrics for Random Forest
rf_report = classification_report(y_test, y_pred_rf, target_names=label_encoder.classes_)
rf_accuracy = accuracy_score(y_test, y_pred_rf)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
```

```python
y_train_dnn = to_categorical(y_train)
y_test_dnn = to_categorical(y_test)  # Convert target to categorical for DNN
```

```python
dnn_model = Sequential([
    Dense(128, input_dim=X_train.shape[1], activation='relu'),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(len(label_encoder.classes_), activation='softmax')
])
```

```python
# Compile the DNN model
dnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the DNN model
dnn_history = dnn_model.fit(X_train, y_train_dnn, epochs=10, batch_size=64, validation_split=0.2, verbose=0)
```

```python
# Evaluate the DNN model
dnn_loss, dnn_accuracy = dnn_model.evaluate(X_test, y_test_dnn, verbose=0)
y_pred_dnn = np.argmax(dnn_model.predict(X_test), axis=1)
```

```python
# Metrics for DNN
dnn_report = classification_report(y_test, y_pred_dnn, target_names=label_encoder.classes_)
```

```python
print(f'Accuracy : ' , rf_accuracy)
print(f'Classfication_report :\n', rf_report)
```
